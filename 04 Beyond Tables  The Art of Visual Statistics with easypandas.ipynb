````python
from Libaldomath import *
from easypandas import *
````

````python
P=easypandas('easydata.xlsx')  
````

## 🛡️ The Philosophy: Data is a Battlefield
We don't just "plot" data. We analyze it in two strategic rounds:
1.  **The Scouting:** Analyzing the individual behavior of a variable (The Fighters).
2.  **The Clash:** Facing two variables against each other to see who dominates the trend (`plot_Versus`).

---

## 🚀 1. The Power of `gfield` Integration
Forget about `KeyError`. Thanks to the `gfield` engine, `easypandas` is now **case-insensitive**. Whether you type `AGE`, `age`, or `AgE`, the library understands your intent.

```python
# The "Lazy but Smart" way to plot
P.versusplot('AGE', 'FACTOR', color='SEX') 
```
*Note: Even if your Excel says "Age" and "Sex", easypandas handles the translation for you.*

---

````python
P.versusplot('Age', 'Factor', color='Sex')
````

````python
P.versusplot('AGE', 'Factor', color='Sex') # testing AGE vs Age
````

## 🥊 2. Round 1: Individual Profiling (`familystat`)
Before the fight, we need to know who is entering the ring. For categorical data like `COLOR` or `SEX`, we use `familystat`.

- **Visual Table:** A "Leaderboard" of frequencies.
- **The Duo-Plot:** A Bar Chart for quantity and a Pie Chart for distribution side-by-side.

```python
P.familystat('Color')
```

````python
P.familystat('Color')
````

## ⚔️ 3. See all in One (`generalplot`)
This is where the magic happens. By using `generalplot`, we generate a multi-dimensional scatter plot.

````python
P.generalplot()
````

### Why it's "Beautiful but Powerful":
* **X & Y Axis:** The main battlefield.
* **Color (The 3rd Dimension):** We use the `color` parameter to divide the fighters into groups (e.g., by Gender or Category).
* **Auto-Labeling:** Title and axis labels are generated automatically using the correct "Real Name" from your data.
