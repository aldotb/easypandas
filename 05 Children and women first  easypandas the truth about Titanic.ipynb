```python
from easypandas import *
P=easypandas("titanic.csv")
%pprint # force jupyter render data list horizontal
```

```python
P.summary()
```

```python
P.fillnan('Age', 28) # prudent value promedio
```

```python
P.superedit()
```

```python
#chek again
P.summary()
```

```python
P.fillnan('Cabin', 'X') # no important for analisis
```

```python
# last chek
P.summary()
```

```python
P.insertcolumn('Category','Adult') 
```

```python
field="Category" 
value=["Child","Teen","Senior"]  
condit=["Age<13","Age>12 and Age<18","Age>60"]
P.multiset(field,value,condit)
```

```python
# last chek
P.summary()
```

```python
# 1. Renombramos para que sea más profesional
P.rename('Ticket', 'Tcode') # Ojo: asegúrate de que el original sea 'Ticket' o 'Tiket'
P.rename('Fare', 'Cost')

# 2. Insertamos la columna vacía al principio
P.insertcolumn('Tiket', 0)

# 3. Configuramos el Multiset con los nombres corregidos
field = "Tiket"
labels = ["D", "C", "B", "A"]

# Importante: Todo ahora es 'Cost'
condit = [
    "Cost < 11",
    "Cost > 10.9 and Cost < 31",
    "Cost > 30.9 and Cost < 101",
    "Cost > 100.9"
]

# Usamos 'labels' en lugar de 'value'
P.multiset(field, labels, condit)
```

```python
# last chek
P.summary()
```

```python
P.familystat('Tiket')
```

```python
P.qty()
```

```python
P.familystat('Survived')
```

```python
# Cruzamos la riqueza (Tiket) contra la Supervivencia (Survived)
P.datagrid(
    col_filters=["Survived == 1", "Survived == 0"], 
    row_filters=["Tiket == 'A' or Tiket == 'B'", "Tiket == 'D' or Tiket == 'C'"],
    col_names=["Live", "Dead"],
    row_names=["Rich (A,B)", "Poor (C,D)"], 
    title=" Live or Dead between Class A,B Vs Class C,D"
    
)
```

```python
# Cruzamos la riqueza (Tiket) contra la Supervivencia (Survived)
P.datagrid(
    col_filters=["Survived == 1", "Survived == 0"], 
    row_filters=["(Tiket == 'A' or Tiket == 'B') and Category=='Child'", "(Tiket == 'D' or Tiket == 'C') and Category=='Child'"],
    col_names=["Live", "Dead"],
    row_names=["Child Rich (A,B)", "Child Poor (C,D)"], 
    title=" Live or Dead between Child in  Class A,B Vs Class C,D"
    
)
```
