```python
field="Category" 
value=["Child","Teen","Senior"]  
condit=["Age<13","Age>12 and Age<18","Age>60"]
P.multiset(field,value,condit)
```

```python
# 1. Renombramos para que sea más profesional
P.rename('Ticket', 'Tcode') # Ojo: asegúrate de que el original sea 'Ticket' o 'Tiket'
P.rename('Fare', 'Cost')

# 2. Insertamos la columna vacía al principio
P.insertcolumn('Tiket', 0)

# 3. Configuramos el Multiset con los nombres corregidos
field = "Tiket"
labels = ["D", "C", "B", "A"]

v# Importante: Todo ahora es 'Cost'
condit = [
    "Cost < 11",
    "Cost > 10.9 and Cost < 31",
    "Cost > 30.9 and Cost < 101",
    "Cost > 100.9"
]

# Usamos 'labels' en lugar de 'value'
P.multiset(field, labels, condit)
```

```python
# last chek
P.summary()
```
