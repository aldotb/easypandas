# easypandas — Simple Data Exploration for Students

`easypandas` is a small educational wrapper around **pandas** that provides a simpler interface for beginners learning data analysis.

The goal is to allow students to explore datasets using simple and intuitive commands.

Internally it uses the power of **pandas**, but exposes a friendlier syntax.

### More examples 

```python
# ============================================
# 1. Cyty with name start whith vocal 
# ============================================
# pandas: bios[bios['born_city'].str.contains(r'^[AEIOUaeiou]', na=False)]
# easypandas: 
B.where("born_city.str.contains('^[AEIOUaeiou]', na=False)")
```

```python
# ============================================
# 2. Name including two vocals
# ============================================
# pandas: bios[bios['name'].str.contains(r'^[^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*$', na=False)]
# easypandas:
B.where("name.str.contains('^[^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*$', na=False)")
```

```python
# ============================================
# 3. Names tahe end whith  'son' or 'sen'
# ============================================
# pandas: bios[bios['name'].str.contains(r'son$|sen$', case=False, na=False)]
# easypandas:
B.where("name.str.contains('son$|sen$', case=False, na=False)")
```

```python
# ============================================
# 4. Date wthat start whith '19'
# ============================================
# pandas: bios[bios['born_date'].str.contains(r'^19', na=False)]
# easypandas:
B.where("born_date.str.contains('^19', na=False)")
```

```python
# ============================================
# 5. Names whitout vocals
# ============================================
# pandas: bios[bios['name'].str.contains(r'^[^AEIOUaeiou]*$', na=False)]
# easypandas:
B.where("name.str.contains('^[^AEIOUaeiou]*$', na=False)")
```

```python
# ============================================
# 6. Names with a hyphen or apostrophe
# ============================================
# pandas: bios[bios['name'].str.contains(r"[-']", na=False)]
# easypandas:
B.where("name.str.contains(\"[-']\", na=False)")
```

```python
# ============================================
# 7. Cities with exactly 7 characters
# ============================================
# pandas: bios[bios['born_city'].str.contains(r'^.{7}$', na=False)]
# easypandas:
B.where("born_city.str.contains('^.{7}$', na=False)")
```

```python
# ============================================
# 8. Names with 3 or more vowels
# ============================================
# pandas: bios[bios['name'].str.contains(r'([AEIOUaeiou].*){3,}', na=False)]
# easypandas:
B.where("name.str.contains('([AEIOUaeiou].*){3,}', na=False)")
```

## Adding / Removing Columns

---
##### Load data
---

```python
# pandas:
coffee = pd.read_csv('./warmup-data/coffee.csv')

# easypandas:
C = easypandas('./warmup-data/coffee.csv')  # ONE LINE!
```

---
##### ASSIGN VALUES (avoiding the warning)
---

```python
# pandas:
coffee['price'] = 4.99

# or better:
C.pdobj['price'] = 4.99  # If you prefer direct pandas
```

---
##### DROP COLUMNS
---

```python
# pandas:
coffee.drop(columns=['price'], inplace=True)

# easypandas:
C.drop("price")  # Direct method!
```

---
##### RENAME COLUMNS
---

```python
# pandas:
coffee.rename(columns={'new_price': 'price'}, inplace=True)

# easypandas:
C.rename("new_price", "price")  # Simple!
```

---
##### COPY DATAFRAME
---

```python
# pandas:
bios_new = bios.copy()

# easypandas:
B_new = easypandas(B.pdobj)  # Creates an independent copy
```

---
##### EXTRACT FIRST NAME (split)
---

```python
# pandas:
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]

# easypandas:
B_new.pdobj['first_name'] = B_new['name'].str.split(' ').str[0]
```

---
##### QUERY (FILTER)
---

```python
# pandas:
bios_new.query('first_name == "Keith"')

# easypandas:
B_new.where('first_name == "Keith"')
```

---
##### CONVERT TO DATETIME
---

```python
# pandas:
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'])

# easypandas:
B_new.pdobj['born_datetime'] = pd.to_datetime(B_new['born_date'])
```

---
##### EXTRACT YEAR
---

```python
# pandas:
bios_new['born_year'] = bios_new['born_datetime'].dt.year

# easypandas:
B_new.pdobj['born_year'] = B_new['born_datetime'].dt.year
```

---
##### SELECT COLUMNS
---

---
##### SAVE TO CSV
---

```python
# pandas:
bios_new.to_csv('./data/bios_new.csv', index=False)

# easypandas:
B_new.pdobj.to_csv('./data/bios_new.csv', index=False)
```
